```python
import pandas as pd
import numpy as np
import os 
```

```python
path = '.\Twitter-repo\CSE243-TweetResponsePrediction\Data-dirty'
files = os.listdir(path)
```

```python
df = pd.DataFrame()
for name in files:
    tmp = pd.read_csv(os.path.join(path, name))
    df = df.append(tmp)
```

```python
df.dropna(subset=['tweets_Date'], inplace=True)
df = df.drop(labels=['listing_value', 'tweets_Name', 'remove'], axis=1).reset_index(drop=True)
df.to_csv('.\Twitter-repo\CSE243-TweetResponsePrediction\Data-clean\Jan-Oct-2020.csv', index=False)
```
